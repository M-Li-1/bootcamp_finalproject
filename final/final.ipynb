```python
import imageio.v3 as iio
import matplotlib.pyplot as plt 
import numpy as np

import skimage as ski
 
from skimage.transform import resize 

from scipy import stats, ndimage

import os 

#1. Import a 500 x 500 pixel PNG image (8bit, greyscale). Then, show the image.
import imageio.v3 as iio
import matplotlib.pyplot as plt 
import numpy as np
import skimage as ski
from skimage.transform import resize 
from scipy import stats, ndimage
import os 


original_im = None


while True:
    image_file_name = input("Enter file name (only grayscale PNG image): ")
    if image_file_name.strip().lower() == 'quit':  
        break
    
    image_path_input = input("Enter path to your PNG image file: ").strip().strip('"')
    if image_path_input.strip().lower() == 'quit':  
        break
    
    
    image_path = os.path.join(image_path_input, image_file_name)
    
    if os.path.exists(image_path):
        try:
            original_im = iio.imread(image_path, mode="L")
            print(f"Loaded image: {os.path.basename(image_path)} ({original_im.shape[1]}x{original_im.shape[0]} pixels)")
            break
        except Exception as e:
            print(f"Error loading image: {str(e)}")
    else:
        print(f"File not found: {image_path}")


if original_im is not None: #corrected by deepseek to quit the loop without error
  
    fig, ax = plt.subplots()
    ax.imshow(original_im, cmap="gray")
    plt.title("Original Image")
    plt.show()

 
    blurred_im = ndimage.gaussian_filter(original_im.astype(float), sigma=4)
    blurred_im = np.clip(blurred_im, 0, 255).astype(np.uint8)
    
    
    fig, ax = plt.subplots()
    ax.imshow(blurred_im, cmap="gray")
    plt.title("Blurred Image")
    plt.show()
    
   
    
else:
    print("Program terminated by user or no valid image loaded.")

#3. calculate the pixel original and blurred image. This was generated by Deepseek. 
if original_im is not None:
    plt.figure(figsize=(10, 6))
    plt.hist(original_im.ravel(), bins=256, alpha=0.5, label='Original', color='blue')
    plt.hist(blurred_im.ravel(), bins=256, alpha=0.5, label='Blurred', color='red')
    plt.title('Grayscale Histogram Comparison')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xlim(0, 255)
    #plt.savefig('histograms.png', dpi=150)
    plt.show()

    # Statistical comparison using KS test
    ks_stat, p_value = stats.ks_2samp(original_im.ravel(), blurred_im.ravel())
    print(f"Kolmogorov-Smirnov Test Results:")
    print(f"Statistic = {ks_stat:.5f}, p-value = {p_value:.5f}")

    if p_value < 0.05:
        print("Conclusion: Significant difference between distributions (p < 0.05)")
    else:
        print("Conclusion: No significant difference between distributions (p >= 0.05)")
    #5. Output the blurred image as PNG in 250 x 250 pixel. 
    blurred_im_resized = resize(blurred_im, (250, 250), anti_aliasing=True)

    fig, ax = plt.subplots()
    ax.imshow(blurred_im_resized, cmap="gray")

    blurred_im_resized_8bit = (blurred_im_resized * 255).astype(np.uint8) #correted by Deepseek to ensure output in 8-bit. 
    iio.imwrite('blurred_resized.png', blurred_im_resized_8bit)
```
